```python
summary_stats = performance[['math_score','reading_score','writing_score']].describe()
print(f'Summary Statistics for Numerical Variables:\n{summary_stats}')
```

```python
categorical_freq = {
    'gender': performance['gender'].value_counts(),
    'race_ethnicity': performance['race_ethnicity'].value_counts(),
    'parental_level_of_education': performance['parental_level_of_education'].value_counts(),
    'lunch': performance['lunch'].value_counts(),
    'test_preparation_course': performance['test_preparation_course'].value_counts()
}
print('Frequency of Categories for Categorical Variables:')
for i,freq in categorical_freq.items():
    print(f'\n{freq}')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
ax = axes.flatten()
numerical_vars = ['math_score', 'reading_score', 'writing_score']
for i, var in enumerate(numerical_vars):
    sns.histplot(performance[var], kde=True, ax=ax[i], color='skyblue', bins=20)
    ax[i].set_title(f'Distribution of {var}')
    ax[i].set_xlabel(var)
    ax[i].set_ylabel('Frequency')
categorical_vars = ['gender', 'race_ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course']
for var in categorical_vars:
    sns.countplot(x=var, data=performance, ax=ax[i+1])
    ax[i+1].set_title(f'Distribution of {var}')
    ax[i+1].set_xlabel(var)
    ax[i+1].set_ylabel('Count')
    ax[i+1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis: Explore the relationship between numerical variables
# Scatter plot matrix (pair plot) for numerical variables
sns.pairplot(performance[['math_score', 'reading_score', 'writing_score']])
plt.suptitle("Pair Plot of Numerical Variables", y=1.02)
plt.show()

# Compare the average scores between different groups using box plots
# Box plot for math_score by gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='math_score', data=performance)
plt.title("Box Plot of Math Score by Gender")
plt.show()

# Box plot for math_score by race_ethnicity
plt.figure(figsize=(10, 6))
sns.boxplot(x='race_ethnicity', y='math_score', data=performance)
plt.title("Box Plot of Math Score by Race/Ethnicity")
plt.show()

# Box plot for math_score by parental_level_of_education
plt.figure(figsize=(12, 6))
sns.boxplot(x='parental_level_of_education', y='math_score', data=performance)
plt.title("Box Plot of Math Score by Parental Level of Education")
plt.xticks(rotation=45)
plt.show()

# Box plot for math_score by lunch
plt.figure(figsize=(8, 6))
sns.boxplot(x='lunch', y='math_score', data=performance)
plt.title("Box Plot of Math Score by Lunch Type")
plt.show()

# Box plot for math_score by test_preparation_course
plt.figure(figsize=(8, 6))
sns.boxplot(x='test_preparation_course', y='math_score', data=performance)
plt.title("Box Plot of Math Score by Test Preparation Course")
plt.show()
```

```python
correlation_matrix = performance[['math_score', 'reading_score', 'writing_score']].corr()

# Visualize correlations using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap of Scores")
plt.show()
```

```python
# Compare average scores between different gender groups
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='reading_score', data=performance)
plt.title("Average Reading Score by Gender")
plt.show()

# Compare average scores between different race/ethnicity groups
plt.figure(figsize=(10, 6))
sns.barplot(x='race_ethnicity', y='reading_score', data=performance)
plt.title("Average Reading Score by Race/Ethnicity")
plt.show()

# Compare average scores between different parental education levels
plt.figure(figsize=(12, 6))
sns.barplot(x='parental_level_of_education', y='reading_score', data=performance)
plt.title("Average Reading Score by Parental Level of Education")
plt.xticks(rotation=45)
plt.show()

# Compare average scores between different lunch types
plt.figure(figsize=(8, 6))
sns.barplot(x='lunch', y='reading_score', data=performance)
plt.title("Average Reading Score by Lunch Type")
plt.show()

# Compare average scores between different test preparation course completion statuses
plt.figure(figsize=(8, 6))
sns.barplot(x='test_preparation_course', y='reading_score', data=performance)
plt.title("Average Reading Score by Test Preparation Course Completion")
plt.show()
```

## Effect of Course Taken on Marks

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='test_preparation_course', y='math_score', data=performance)
plt.title("Average Math Score by Test Preparation Course Completion")
plt.xlabel("Test Preparation Course Completion")
plt.ylabel("Average Math Score")
plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(x='test_preparation_course', y='reading_score', data=performance)
plt.title("Average Reading Score by Test Preparation Course Completion")
plt.xlabel("Test Preparation Course Completion")
plt.ylabel("Average Reading Score")
plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(x='test_preparation_course', y='writing_score', data=performance)
plt.title("Average Writing Score by Test Preparation Course Completion")
plt.xlabel("Test Preparation Course Completion")
plt.ylabel("Average Writing Score")
plt.show()
```
